```python
import pandas as pd
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt

data = pd.read_excel('https://github.com/bnsreenu/python_for_microscopists/blob/master/other_files/K_Means.xlsx?raw=true')
print(data)
```

```python
kmeans = KMeans(n_clusters = 3, init='k-means++', max_iter=300, n_init=10, random_state=0)
```

```python
model = kmeans.fit(data)
```

```python
pred_val = kmeans.predict(data)
```

```python
plt.scatter(data['X'], data['Y'], c=pred_val, facecolor = 'orange')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, c='blue')
```
